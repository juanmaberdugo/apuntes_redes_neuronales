Segun entendi en la comparacion que hace, teniendo un conjunto de datos lo que se intenta hacer con las redes neuronales es parecido a hacer regresion lineal con un conjunto de datos, osea encontrar la linea de mejor ajuste de los datos, idear una funcion y con los datos de entrada intentar predecir los resultados  de salida.

### Que es ReLU

ReLU es una regla de activacion de una red neuronal, funciona filtrando los datos con los que se entrena la red, por ejemplo tenemos un ReLu tal que:

```python
dato1 = 1
dato2 = -2

def relu(x):
    if x < 0:
        return 0
    else:
        return x

print(relu(dato1))  # Deberia imprimir 1
print(relu(dato2))  # Deberia imprimir 0
```

este relu deberia aceptar los datos que sean mayores mayores o iguales que 0 y devolverlos igual, pero con los valores negativos lo que hace es cambiar el valor a cero, este seria un ejemplo basico de un relu
